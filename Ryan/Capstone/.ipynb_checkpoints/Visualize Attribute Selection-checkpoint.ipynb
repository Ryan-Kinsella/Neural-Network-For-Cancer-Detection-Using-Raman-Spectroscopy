```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("useful_features_1.csv")
```

```python
# Faceting: act of breaking data variables up across multiple subplots 
# and combining those subplots into a single figure.

g = sns.FacetGrid(df, col='class')
g = g.map(sns.kdeplot, '50')

# This finds trends with the class wrt the column. 
# more examples at https://seaborn.pydata.org/generated/seaborn.FacetGrid.html
```

```python
for col in df.columns:
    if col != 'class':
        g = sns.FacetGrid(df, col='class')
        g = g.map(sns.kdeplot, str(col))
```
